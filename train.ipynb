```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("train.csv")
```

```python
data
```

```python
data.isna().sum()
```

```python
x = data.drop("Transported", axis = 1)
y = data["Transported"]
```

```python
y.astype("int8")
```

```python
 x_train, x_test, y_train, y_test = train_test_split( x, y, test_size=0.25, random_state=0)
x_train
```

```python
x_train["HomePlanet"] = x_train["HomePlanet"].fillna(np.random.choice(["Earth", "Europa", "Mars"], p = [0.54, 0.25, 0.21]))
x_train["CryoSleep"] = x_train["CryoSleep"].fillna(np.random.choice([True, False], p = [0.36, 0.64]))
x_train["Destination"] = x_train["Destination"].fillna(np.random.choice(["TRAPPIST-1e", "55 Cancri e", "PSO J318.5-22"], p = [0.7, 0.21, 0.09]))
x_train["Age"] = x_train["Age"].fillna(x_train["Age"].median())
x_train["VIP"] = x_train["VIP"].fillna(np.random.choice([True, False], p = [0.02, 0.98]))
if x_train["CryoSleep"] is True :
    x_train["RoomService"] = x_train["RoomService"].fillna(x_train["RoomService"].median())
    x_train["FoodCourt"] = x_train["FoodCourt"].fillna(x_train["FoodCourt"].median())
    x_train["ShoppingMall"] = x_train["ShoppingMall"].fillna(x_train["ShoppingMall"].median())
    x_train["Spa"] = x_train["Spa"].fillna(x_train["Spa"].median())
    x_train["VRDeck"] = x_train["VRDeck"].fillna(x_train["VRDeck"].median())
else:
    x_train["RoomService"] = x_train["RoomService"].fillna(0)
    x_train["FoodCourt"] = x_train["FoodCourt"].fillna(0)
    x_train["ShoppingMall"] = x_train["ShoppingMall"].fillna(0)
    x_train["Spa"] = x_train["Spa"].fillna(0)
    x_train["VRDeck"] = x_train["VRDeck"].fillna(0)
    
```

```python
x_train["Cabin"] = x_train["Cabin"].fillna(np.random.choice(["G/1025/S", "G/512/P"]))
```

```python
for i in x_train.index:
    x_train["Cabin"][i] = x_train["Cabin"].str.split("/")[i][2]
```

```python
x_train = pd.get_dummies(x_train, columns = ["HomePlanet"], dtype = "int8")
x_train = pd.get_dummies(x_train, columns = ["Destination"], dtype = "int8")
x_train["CryoSleep"] = x_train["CryoSleep"].astype('int8')
```

```python
x_train["Cabin"] = (x_train["Cabin"] == "P").astype('int8')
x_train["VIP"] = x_train["VIP"].astype('int8')

x_train
```

```python
x_train = x_train.drop("Name", axis = 1)
```
